## **Introduction**

the combined influence of economic factors, investor behavior, and random
fluctuations. Traditional deterministic forecasting methods often fail to
capture this uncertainty. To address this, **probabilistic simulation-based
approaches** such as Monte Carlo Simulation are widely used in quantitative
finance and risk management.

This project applies **Monte Carlo Simulation techniques** to model the future

---

## **What is Monte Carlo Simulation?**

Monte Carlo Simulation is a computational technique that uses **random sampling**
to model complex systems and quantify uncertainty. Instead of producing a
single forecast, it generates **thousands of possible future scenarios**
based on assumed statistical properties of the underlying process.

In financial modeling, Monte Carlo Simulation is commonly used to:
- Model asset price dynamics
- Estimate risk measures such as Value-at-Risk (VaR)
- Analyze uncertainty in long-term forecasts
- Perform scenario and stress analysis

---

## **Why Use Monte Carlo Simulation for Financial Markets?**

Financial time series exhibit randomness, volatility clustering, and
non-deterministic behavior. Monte Carlo Simulation is particularly suitable
because it:

- Produces a **distribution of future outcomes** rather than a single point estimate
- Explicitly models **uncertainty and risk**
- Allows comparison between different modeling assumptions
- Is widely used in industry for pricing, forecasting, and risk management

In this project, Monte Carlo Simulation enables us to analyze not only the
expected future value of the NIFTY BANK Index but also the **range of possible
outcomes and associated downside risks**.

## **Monte Carlo Simulation**

```python
# ---------- settings ----------
SEED = 42
np.random.seed(SEED)
N_SIMS = 2000
HORIZON = 252        # days
PLOT_PATHS = 40      # sample paths to display

# ---------- prepare returns & start price ----------
df['LogReturns'] = np.log(df['Close'] / df['Close'].shift(1))
df = df.dropna(subset=['LogReturns']).reset_index(drop=True)
rets = df['LogReturns'].values
S0 = df['Close'].iloc[-1]
mu = rets.mean(); sigma = rets.std(ddof=1)

# ---------- GBM (vectorized) ----------
Z = np.random.normal(size=(HORIZON, N_SIMS))
dt = 1/252
increments = (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z
log_paths = np.vstack([np.zeros(N_SIMS), np.cumsum(increments, axis=0)])
S_gbm = S0 * np.exp(log_paths)        # shape (HORIZON+1, N_SIMS)
term_gbm = S_gbm[-1, :]



# ---------- VaR & CVaR on log-return losses (95%) ----------
def var_cvar(terminals, S0, alpha=0.95):
    losses = -np.log(terminals / S0)
    VaR = np.percentile(losses, alpha*100)
    CVaR = losses[losses >= VaR].mean()
    return VaR, CVaR

VaR_g, CVaR_g = var_cvar(term_gbm, S0)


print(f"S0={S0:.2f}  mu={mu:.6f}  sigma={sigma:.6f}")
print(f"GBM VaR95={VaR_g:.6f}, CVaR95={CVaR_g:.6f}")

# -------GBM Chart-----------
p = np.percentile(S_gbm, [5,25,50,75,95], axis=1)
x = np.arange(HORIZON+1)

plt.figure(figsize=(13,9))


cmap = plt.cm.ocean_r


for i in range(min(PLOT_PATHS, N_SIMS)):
    plt.plot(x, S_gbm[:, i], color=cmap(i / PLOT_PATHS), alpha=0.6, linewidth=1)

# Median & confidence bands
plt.plot(x, p[2], color='black', linewidth=2, label='Median')
plt.fill_between(x, p[0], p[4], alpha=0.08, color='blue')

plt.title("Monte Carlo Simulation using Geometric Brownian Motion")
plt.xlabel("Days")
plt.ylabel("Index Value")
plt.tight_layout()
plt.savefig("Monte_Carlo_GBM_Chart.svg", bbox_inches="tight")
plt.show()
```

Insights and Interpretation

- The widening fan of simulated paths shows that forecast uncertainty grows rapidly with time, indicating that long-term index levels cannot be predicted precisely and must be assessed probabilistically.
- Although the median trajectory remains close to the current index level, the wide spread of paths highlights substantial upside and downside risk, emphasizing that volatilityâ€”not average returnâ€”is the primary driver of future outcomes.


## **Non - Parametric Bootstrap Simulation**

Unlike Geometric Brownian Motion, which assumes normally distributed returns
and constant volatility, the non-parametric bootstrap approach directly
resamples historical returns.

This method:
- Preserves empirical distribution characteristics
- Captures skewness and heavy tails
- Avoids strong parametric assumptions

```python
# Parameters
N_SIMS = 2000
HORIZON = 252      # 1-year forecast
SEED = 42
np.random.seed(SEED)

term_boot = np.empty(N_SIMS)
for i in range(N_SIMS):
  samp = np.random.choice(rets, size=HORIZON, replace=True)
  term_boot[i] = S0 * np.exp(samp.sum())
```

###**Bootstrap Terminal Distribution Plot**

```python
plt.figure(figsize=(13, 10))
plt.hist(term_boot, bins=60, density=True, alpha=0.8, color='tab:orange')
plt.axvline(S0, color='black', linestyle='--', label='Current Price')
plt.xlabel("Terminal Index Value")
plt.ylabel("Density")
plt.title("Bootstrap Monte Carlo â€” Terminal Distribution (1 Year)")
plt.legend()
plt.tight_layout()
plt.savefig("Bootstrap_Terminal_Distribution.svg", bbox_inches="tight")
plt.show()
```

Insights and Interpretation

- The wide and uneven shape of the distribution shows that **future index values
  can vary a lot**, with some scenarios leading to very large gains or losses.
- Compared to the GBM model, the bootstrap method shows **higher downside risk**
  because it uses real historical returns instead of assuming a normal pattern.


### **Risk Metrics from Bootstrap Simulation**

```python
VaR_b, CVaR_b = var_cvar(term_boot, S0)

print(f"Bootstrap VaR (95%):  {VaR_b:.6f}")
print(f"Bootstrap CVaR (95%): {CVaR_b:.6f}")
```

Risk Metrics from Bootstrap Simulation â€” Simple Interpretation

- A **95% Bootstrap VaR of ~26.7%** means that, under most market conditions,
  losses are expected to remain below about **27% over one year**, but worse
  outcomes are still possible.
- A **95% Bootstrap CVaR of ~36.1%** indicates that **when losses exceed the VaR
  threshold**, the average loss can be around **36%**, highlighting significant
  tail risk.
